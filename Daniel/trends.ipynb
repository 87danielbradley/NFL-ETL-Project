```python
offense= pd.read_csv("roster/offense.csv")
defense= pd.read_csv("roster/defense.csv")
special= pd.read_csv("roster/special.csv")
```

```python
offense=offense.drop(columns=['Unnamed: 0'])
defense=defense.drop(columns=['Unnamed: 0'])
special=special.drop(columns=['Unnamed: 0'])

offense["Name"]=offense["Name"].str.replace('\d+','')
defense["Name"]=defense["Name"].str.replace('\d+','')
special["Name"]=special["Name"].str.replace('\d+','')
```

```python
all_players =pd.DataFrame({})
all_players = pd.concat([offense,defense])
all_players = pd.concat([all_players,special])
```

```python
for player in list(all_players["Name"]):
    #name = player.replace(" ","%20")
    #url = f'https://trends.google.com/trends/explore?q={name}&geo=US'
    pytrends = TrendReq(hl='en-US', tz=360, kw_list=[player])
```
